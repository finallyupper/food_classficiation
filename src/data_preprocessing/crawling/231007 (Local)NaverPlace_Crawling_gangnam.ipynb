```python
# frame 변경 메소드
def switch_frame(frame):
    driver.switch_to.default_content()  # frame 초기화
    driver.switch_to.frame(frame)  # frame 변경
    res
    soup

# 페이지 다운
def page_down(num):
    body = driver.find_element(By.CSS_SELECTOR, 'body')
    body.click()
    for i in range(num):
        body.send_keys(Keys.PAGE_DOWN)
```

```python
#store_id = 0
switch_frame('searchIframe')
page_down(40) ###
driver.implicitly_wait(5)
a = 29 ######
b=51
for i in tqdm(range(a, b)):
    # 가게 들어가기
    try:
        driver.find_element(By.XPATH, "/html/body/div[3]/div/div[3]/div[1]/ul/li["+str(i)+"]/div[1]/a[1]/div/div/span[1]").click()
    except: 
        try:
            time.sleep(3)
            driver.find_element(By.XPATH, "/html/body/div[3]/div/div[3]/div[1]/ul/li["+str(i)+"]/div[1]/a[1]/div/div/span[1]").click()
        except: continue
    time.sleep(4)
    switch_frame('entryIframe')
    time.sleep(5)
    # 가게명
    storename = driver.find_element(By.CSS_SELECTOR, '#_title > span.Fc1rA').text


    # 분류
    storetype = driver.find_element(By.XPATH, '//*[@id="_title"]/span[2]').text
    # 별점(5점 만점)
    try:
        star = driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[2]/div[1]/div[2]/span[1]/em').text
    except: star = np.NaN
    # 주소
    try: address = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div/div[6]/div/div[2]/div/div/div[1]/div/a/span[1]').text
    except: address = np.NaN
    # 전화번호
    try: number = driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[7]/div/div[2]/div/div/div[4]/div/span[1]').text
    except: number = np.NaN
    # 메뉴 개수
    # page_down(1)
    time.sleep(1)
    try: num_of_menu = driver.find_element(By.CLASS_NAME,"place_section_count").text
    except: continue

    # 가게 id
    store_id += 1
    # 메뉴명 & 메뉴 이미지

    if "메뉴" in driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[3]/span').text: 
        driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[3]/span').click()
    elif "메뉴" in driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[2]/span').text:
        driver.find_element(By.XPATH,'//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[2]/span').click() 
    elif "메뉴" in driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[3]/span').text:
        driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[5]/div/div/div/div/a[3]/span').click()
    else:
        driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div/header/div[2]/div/a[3]/span').click()
    
    print("메뉴 진입 성공!>")
    driver.implicitly_wait(3)
    for m in range(int(num_of_menu)):#num_of_menu
        try:
            try:
                menu = driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[7]/div[2]/div[1]/div[1]/ul/li[' + str(m+1) +']/a/div[2]/div[1]/div/span').text
            except:
                time.sleep(3)
                try: menu = driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[7]/div[2]/div[1]/div/ul/li[' + str(m+1) +']/a/div[2]/div[1]/div/span[1]').text
                except:
                    time.sleep(3)
                    menu = driver.find_element(By.XPATH, '//*[@id="root"]/div[3]/div/div/div[2]/div[2]/div[1]/ul/li[' + str(m+1) +']/div/a[1]/div[2]/div[1]').text
            
            #menu = driver.find_element(By.XPATH, '//*[@id="root"]/div[3]/div/div/div[2]/div[2]/div/ul/li['+str(m+1)+']/a/div[2]/div[1]/div/span').text
            #print(menu) ###
            img_url = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div/div[7]/div[2]/div[1]/div/ul/li["+str(m+1)+"]/a/div[1]/div/div")
            img = img_url.get_attribute('style')
            img = img.split('url(')[-1]
            img = img.split(')')[0]
            try:
                urlretrieve(img[1:-1], img_folder + f'/{str(store_id)}_{menu}.jpg')
            except: continue
            d = dict()
            d['가게id'] = store_id
            d['가게명'] = storename
            d['분류'] = storetype
            d['별점'] = star
            print('가게=',store_id) ###
            d['주소'] = address
            d['전화번호'] = number
            d['# 메뉴'] = num_of_menu
            d['메뉴명'] = menu
            d['이미지파일명'] = str(store_id) + '_' + menu +'.jpg'
            print(d)
            data.append(d)
        except: 
            print('------------------------exception in', storename)
            break
    switch_frame('searchIframe')
    driver.implicitly_wait(4)
```
