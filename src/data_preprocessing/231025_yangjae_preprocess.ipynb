```python
# 딕셔너리로 변환 (콤마(,) 기준 구분, 띄어쓰기 제거, outlier은 손수 고쳤음)
categories = list(sample_list_df['카테고리'])
dic = dict()

for (i, cat) in enumerate(categories):
    details = [item.strip() for item in sample_list_df.iloc[i]['세부사항'].split(',')]
    dic[cat] = details

lst = ['쌀밥,기타잡곡밥,콩밥,보리밥,돌솥밥,현미밥,흑미밥,감자밥,곤드레밥,김치볶음밥,주먹밥,볶음밥,일반 비빔밥,전주비빔밥,삼선볶음밥,새우볶음밥,알밥,산채비빔밥,오므라이스,육회비빔밥,해물볶음밥,열무 비빔밥,불고기덮밥,소고기국밥,송이덮밥,오징어덮밥,자장밥,잡채밥,잡탕밥,장어덮밥,제육덮밥,짬뽕 밥,순대국밥,카레라이스,전주콩나물국밥,해물덮밥,회덮밥,소머리국밥,돼지국밥,하이라이스,김치김 밥,농어초밥,문어초밥,새우초밥,새우튀김롤,샐러드김밥,광어초밥,소고기김밥,갈비삼각김밥,연어롤, 연어초밥,유부초밥,장어초밥,참치김밥,참치마요삼각김밥,치즈김밥,캘리포니아롤,한치초밥,일반김밥']
_details = [item.strip() for item in lst[0].split(',')]
dic['밥류'] = _details

print(dic)
```

```python
# 새로 담을 데이터프레임
new_df = pd.DataFrame(columns= ['Unnamed: 0','가게id','가게명','분류','별점','주소','전화번호','# 메뉴','메뉴명','이미지파일명','카테고리', '수정메뉴명'])
```

```python
except_lst_yangjae = pd.DataFrame(columns= yangjae.columns) #세부사항에 없는 샘플들의 인덱스
yangjae_df = pd.DataFrame(yangjae)

cnt = 1
start_time = time.time()

for idx in tqdm(range(len(yangjae_df))):
    condition = False ## new_df에 들어갔는지 검사
    sample = yangjae_df.iloc[[idx]]
    sample_menu_name = yangjae.iloc[idx]['메뉴명']
    for cat_name in dic.keys():     
        for menu_name in dic[cat_name]:
            if menu_name in sample_menu_name:# sample_menu_name in menu_name: #음식리스트_샘플의 카테고리->디테일의 음식명을 포함하고 있으면 이에 삽입함. 
                condition = True ##
                sample = sample.copy()
                sample['카테고리'] = cat_name #cat_name
                sample['수정메뉴명'] = menu_name
                new_df = pd.concat([new_df, sample], ignore_index = True)
                cnt+=1
                break
                
    if condition is False: #안들어갔으면 추가
        except_lst_yangjae= pd.concat([except_lst_yangjae, sample], ignore_index = True)

end_time = time.time()
total_time = end_time - start_time
print(f"Total time taken: {total_time} seconds")
```

```python
new_df2 = pd.DataFrame(columns= ['Unnamed: 0','가게id','가게명','분류','별점','주소','전화번호','# 메뉴','메뉴명','이미지파일명','카테고리', '수정메뉴명'])
still_except_lst_yangjae = pd.DataFrame(columns= except_lst_yangjae.columns) #세부사항에 없는 샘플들의 인덱스

cnt = 1
start_time = time.time()

for idx in tqdm(range(len(except_lst_yangjae))):
    condition = False ## new_df2에 들어갔는지 검사
    sample = except_lst_yangjae.iloc[[idx]]
    sample_menu_name = except_lst_yangjae.iloc[[idx]]['메뉴명'][idx] ############3
    for (i, cat_name) in enumerate(dic.keys()):     
        for menu_name in dic[cat_name]:
            if key_cats[i] in sample_menu_name: ### 수정한 부분 ##########3
                condition = True ##
                sample = sample.copy()
                sample['카테고리'] = cat_name #cat_name
                sample['수정메뉴명'] = menu_name
                new_df2 = pd.concat([new_df2, sample], ignore_index = True)
                cnt+=1
                break 
                
    if condition is False: #안들어갔으면 추가
        still_except_lst_yangjae= pd.concat([still_except_lst_yangjae, sample], ignore_index = True)

end_time = time.time()
total_time = end_time - start_time
print(f"Total time taken: {total_time} seconds")
```
